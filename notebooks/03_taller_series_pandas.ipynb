````python
# Crea dos Series numéricas y súmalas.

# Crear dos Series numéricas
serie1 = pd.Series([10, 20, 30, 40])
serie2 = pd.Series([1, 2, 3, 4])

# Sumarlas
resultado = serie1 + serie2

print("Serie 1:")
print(serie1)
print("\nSerie 2:")
print(serie2)
print("\nSuma:")
print(resultado)
````

````python
# Resta una Serie de otra

# Usando las Series del ejemplo anterior
serie1 = pd.Series([10, 20, 30, 40])
serie2 = pd.Series([1, 2, 3, 4])

# Resta: serie1 - serie2
resultado_resta = serie1 - serie2

print("Serie 1:")
print(serie1)
print("\nSerie 2:")
print(serie2)
print("\nResta (serie1 - serie2):")
print(resultado_resta)
````

````python
# Multiplica todos los valores de una Serie por 2

# Serie original
serie = pd.Series([10, 20, 30, 40])

# Multiplicar TODOS los valores por 2
serie_x2 = serie * 2

print("Serie original:")
print(serie)
print("\nMultiplicada por 2:")
print(serie_x2)
````

````python
# Divide los valores de una Serie entre otro número

# Serie original
serie = pd.Series([20, 40, 60, 80])

# Dividir todos los valores entre 4
serie_dividida = serie / 4

print("Serie original:")
print(serie)
print("\nDividida entre 4:")
print(serie_dividida)
````

````python
# Realiza una operación combinada: (serie1 + serie2) / 2

# Dos Series numéricas
serie1 = pd.Series([10, 20, 30, 40])
serie2 = pd.Series([1, 2, 3, 4])

# Operación combinada: (serie1 + serie2) / 2
promedio = (serie1 + serie2) / 2

print("Serie 1:")
print(serie1)
print("\nSerie 2:")
print(serie2)
print("\n(serie1 + serie2) / 2:")
print(promedio)
````

````python
# Aplica una función lambda para elevar al cuadrado cada elemento

# Serie de números
serie = pd.Series([2, 4, 6, 8])

print("Serie original:")
print(serie)
print("\nElevada al cuadrado con lambda:")
print(serie.apply(lambda x: x ** 2))
````

````python
# Usa map() para transformar una Serie de nombres a mayúsculas

# Serie de nombres
nombres = pd.Series(['ana', 'luis', 'maría', 'pedro'])

print("Nombres originales:")
print(nombres)
print("\nNombres en mayúsculas con map():")
print(nombres.map(str.upper))
````

````python
# Aplica apply(np.sqrt) a una Serie numérica

# Serie numérica
serie = pd.Series([4, 9, 16, 25, 36])

print("Serie original:")
print(serie)
print("\nRaíz cuadrada con apply(np.sqrt):")
print(serie.apply(np.sqrt))
````

````python
# Crea una función personalizada para clasificar valores y aplícala con apply()

# Función personalizada para clasificar valores
def clasificar(valor):
    if valor < 50:
        return 'Bajo'
    elif valor < 80:
        return 'Medio'
    else:
        return 'Alto'

# Serie de valores
valores = pd.Series([45, 72, 85, 30, 92, 65])

print("Valores originales:")
print(valores)
print("\nClasificación con apply():")
print(valores.apply(clasificar))
````

````python
# Usa map() con un diccionario de reemplazo

# Diccionario de reemplazo
reemplazo = {'junior': 'Empleado', 'senior': 'Líder', 'manager': 'Jefe'}

# Serie original
cargos = pd.Series(['junior', 'senior', 'junior', 'manager', 'senior'])

print("Cargos originales:")
print(cargos)
print("\nReemplazados con map():")
print(cargos.map(reemplazo))
````

````python
# Ordena una Serie de números de forma ascendente

# Serie desordenada
numeros = pd.Series([45, 12, 78, 3, 91, 25])

print("Serie original:")
print(numeros)
print("\nOrdenada ascendente:")
print(numeros.sort_values())
````

````python
# Ordena una Serie de texto alfabéticamente

# Serie de texto desordenada
ciudades = pd.Series(['Zona 5', 'Madrid', 'Ávila', 'Barcelona', 'Sevilla'])

print("Ciudades originales:")
print(ciudades)
print("\nOrdenadas alfabéticamente:")
print(ciudades.sort_values())
````

````python
# Usa sort_index() para ordenar por índice

# Serie con índice desordenado
serie = pd.Series([10, 30, 20, 40], index=['c', 'a', 'd', 'b'])

print("Serie original:")
print(serie)
print("\nOrdenada por índice:")
print(serie.sort_index())
````

````python
# Obtén los tres valores más altos (nlargest)

# Serie de números
ventas = pd.Series([45, 120, 78, 200, 95, 150])

print("Serie original:")
print(ventas)
print("\nTres valores más altos (nlargest):")
print(ventas.nlargest(3))
````

````python
# Obtén los tres valores más bajos (nsmallest)

# Serie de números
ventas = pd.Series([45, 120, 78, 200, 95, 150, 10, 85])

print("Serie original:")
print(ventas)
print("\nTres valores más bajos (nsmallest):")
print(ventas.nsmallest(3))
````

````python
# Filtra una Serie para mostrar solo los valores mayores que 50

# Serie de números
valores = pd.Series([45, 72, 30, 85, 25, 92, 60])

print("Serie original:")
print(valores)
print("\nValores > 50:")
print(valores[valores > 50])
````

````python
# Muestra los valores pares de una Serie

# Serie de números
numeros = pd.Series([1, 4, 7, 12, 9, 16, 23, 18])

print("Serie original:")
print(numeros)
print("\nSolo valores pares:")
print(numeros[numeros % 2 == 0])
````

````python
# Filtra los valores que contengan una palabra específica en Series de texto.

# Serie de texto
productos = pd.Series([
    'Laptop gaming RGB', 
    'Mouse óptico', 
    'Teclado mecánico', 
    'Mouse gaming RGB', 
    'Monitor 24 pulgadas'
])

print("Productos originales:")
print(productos)
print("\nProductos que contienen 'RGB':")
print(productos[productos.str.contains('RGB', case=False)])
````

````python
# Usa una condición doble (por ejemplo, >30 & <70)

# Serie de números
temperaturas = pd.Series([25, 45, 65, 85, 35, 55, 75, 95, 15])

print("Temperaturas originales:")
print(temperaturas)
print("\nTemperaturas entre 30 y 70:")
print(temperaturas[(temperaturas > 30) & (temperaturas < 70)])
````

````python
# Aplica un filtro que devuelva los índices de los valores válidos

# Serie con valores válidos e inválidos (NaN)
serie = pd.Series([10, 25, np.nan, 45, np.nan, 60, 75])

print("Serie original:")
print(serie)
print("\nÍndices de valores válidos:")
print(serie.notna()[serie.notna()].index.tolist())
````

````python
# Crea una Serie con valores NaN y reemplázalos con 0

# Serie con valores NaN
serie = pd.Series([10, np.nan, 30, np.nan, 50, 60, np.nan])

print("Serie original con NaN:")
print(serie)
print("\nNaN reemplazados por 0:")
print(serie.fillna(0))
````

````python
# Usa dropna() para eliminar los nulos

# Serie con valores nulos
serie = pd.Series([10, np.nan, 30, np.nan, 50, 60, np.nan])

print("Serie original con NaN:")
print(serie)
print("\nDespués de dropna():")
print(serie.dropna())
````

````python
# Cuenta los valores nulos con isna().sum()

# Serie con valores nulos
serie = pd.Series([10, np.nan, 30, np.nan, 50, 60, np.nan])

print("Serie original:")
print(serie)
print("\nCantidad de valores nulos:")
print(serie.isna().sum())
````

````python
# Llena los nulos con el promedio de la Serie

# Serie con valores NaN
serie = pd.Series([10, np.nan, 30, np.nan, 50, 60, np.nan])

print("Serie original con NaN:")
print(serie)
print(f"Promedio: {serie.mean():.1f}")
print("\nNaN llenados con promedio:")
print(serie.fillna(serie.mean()))
````

````python
# Reemplaza los nulos con un valor constante

# Serie con valores nulos
serie = pd.Series([10, np.nan, 30, np.nan, 50, 60])

print("Serie original con NaN:")
print(serie)
print("\nNaN reemplazados por valor constante -100:")
print(serie.fillna(-100))
````

````python
# Crea dos Series con índices distintos y súmalas

# Serie 1 con índices específicos
serie1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
# Serie 2 con índices diferentes
serie2 = pd.Series([5, 15], index=['b', 'd'])

print("Serie 1:")
print(serie1)
print("\nSerie 2:")
print(serie2)
print("\nSuma alineada por índices:")
print(serie1 + serie2)
````

````python
# Combina dos Series usando add() con fill_value=0

# Dos Series con índices distintos
serie1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
serie2 = pd.Series([5, 15, 25], index=['b', 'd', 'e'])

print("Serie 1:")
print(serie1)
print("\nSerie 2:")
print(serie2)
print("\nSuma con .add(fill_value=0):")
print(serie1.add(serie2, fill_value=0))
````

````python
# Usa combine_first() para fusionar dos Series

# Serie principal con algunos NaN
serie1 = pd.Series([10, np.nan, 30, np.nan, 50], index=['a', 'b', 'c', 'd', 'e'])
# Serie de respaldo
serie2 = pd.Series([100, 200, 300], index=['b', 'd', 'f'])

print("Serie principal:")
print(serie1)
print("\nSerie respaldo:")
print(serie2)
print("\ncombine_first():")
print(serie1.combine_first(serie2))
````

````python
# Crea Series con índices no coincidentes y observa cómo Pandas alinea los datos

# Series con índices NO coincidentes
serie1 = pd.Series([100, 200, 300], index=['x', 'y', 'z'])
serie2 = pd.Series([10, 20, 30, 40], index=['a', 'x', 'b', 'c'])

print("Serie 1 (índices: x,y,z):")
print(serie1)
print("\nSerie 2 (índices: a,x,b,c):")
print(serie2)
print("\nSuma (Pandas ALINEA por índices):")
print(serie1 + serie2)
````

````python
# Combina Series con operaciones aritméticas complejas

# Series con índices alineados
ventas = pd.Series([100, 150, 200], index=['Ene', 'Feb', 'Mar'])
costos = pd.Series([60, 90, 120], index=['Ene', 'Feb', 'Mar'])

print("Ventas:")
print(ventas)
print("\nCostos:")
print(costos)
print("\nBeneficio (ventas - costos):")
print(ventas - costos)
print("\nMargen % ((ventas - costos)/ventas * 100):")
print(((ventas - costos) / ventas * 100).round(1))
````

````python
# Crea una Serie con índice ['a','b','c','d'] y reindexa a ['a','b','c','d','e']

# Serie original con índices ['a','b','c','d']
serie_original = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

print("Serie original:")
print(serie_original)
print("\nReindex a ['a','b','c','d','e']:")
print(serie_original.reindex(['a', 'b', 'c', 'd', 'e']))
````

````python
# Reindexa una Serie numérica en orden inverso

# Serie numérica original
serie = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

print("Serie original:")
print(serie)
print("\nReindexada en orden inverso:")
print(serie.reindex(['d', 'c', 'b', 'a']))
````

````python
# Usa reindex() con el parámetro fill_value

# Serie original
serie = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

print("Serie original:")
print(serie)
print("\nReindex con fill_value=0:")
print(serie.reindex(['a', 'b', 'c', 'd', 'e'], fill_value=0))
````

````python
# Crea un índice nuevo y rellena los valores faltantes con el promedio

# Serie original
serie = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

print("Serie original:")
print(serie)
print("\nNuevo índice:")
nuevo_indice = ['a', 'b', 'c', 'd', 'e', 'f']
print(nuevo_indice)
print("\nReindex relleno con promedio:")
print(serie.reindex(nuevo_indice, fill_value=serie.mean()))
````

````python
# Usa reindex() con un índice que contenga valores nuevos y existentes

# Serie original
serie = pd.Series([10, 20, 30, 40], index=['a', 'c', 'd', 'f'])

print("Serie original:")
print(serie)
print("\nNuevo índice (existentes + nuevos):")
nuevo_indice = ['a', 'b', 'c', 'd', 'e', 'f']
print(nuevo_indice)
print("\nReindex combinado:")
print(serie.reindex(nuevo_indice))
````

Las Series mejoran la eficiencia mediante operaciones vectorizadas rápidas (sin bucles for), alineación automática por índices y manejo nativo de NaN.
