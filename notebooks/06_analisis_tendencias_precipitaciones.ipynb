## 5. Revisión complementaria de calidad

```python
nulos = pd.DataFrame({
    'nulos': df.isnull().sum(),
    'porcentaje_%': (df.isnull().mean() * 100).round(2)
})
nulos
```

```python
df.duplicated().sum()
```

```python
df['fecha_dt'] = pd.to_datetime(df['fechaobservacion'], errors='coerce')
print("Fechas no convertidas:", df['fecha_dt'].isna().sum())
print("Fecha mínima:", df['fecha_dt'].min())
print("Fecha máxima:", df['fecha_dt'].max())
```

```python
print("Registros con precipitación negativa:", (df['valorobservado'] < 0).sum())
print("Registros con precipitación igual a 0:", (df['valorobservado'] == 0).sum())
print("Porcentaje de precipitación en 0:", round((df['valorobservado'] == 0).mean() * 100, 2), "%")
```

```python
df.duplicated(subset=['codigoestacion', 'codigosensor', 'fechaobservacion']).sum()
```

```python
resumen_consistencia = {
    'codigos_estacion_con_multiples_nombres': int((df.groupby('codigoestacion')['nombreestacion'].nunique() > 1).sum()),
    'codigos_estacion_con_multiples_departamentos': int((df.groupby('codigoestacion')['departamento'].nunique() > 1).sum()),
    'codigos_estacion_con_multiples_municipios': int((df.groupby('codigoestacion')['municipio'].nunique() > 1).sum()),
    'codigos_estacion_con_multiples_zonas': int((df.groupby('codigoestacion')['zonahidrografica'].nunique() > 1).sum()),
    'codigos_estacion_con_multiples_coordenadas': int((df.groupby('codigoestacion')[['latitud','longitud']].nunique().max(axis=1) > 1).sum()),
    'nombres_estacion_reutilizados_en_varios_codigos': int((df.groupby('nombreestacion')['codigoestacion'].nunique() > 1).sum())
}
resumen_consistencia
```

```python
print("Estaciones únicas:", df['codigoestacion'].nunique())
print("Nombres de estación únicos:", df['nombreestacion'].nunique())
print("Departamentos únicos:", df['departamento'].nunique())
print("Municipios únicos:", df['municipio'].nunique())
print("Zonas hidrográficas únicas:", df['zonahidrografica'].nunique())
```
