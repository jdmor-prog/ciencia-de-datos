## Objetivo
Realizar la carga, inspección y diagnóstico inicial del dataset para identificar su estructura, tipos de variables, valores faltantes y posibles inconsistencias usando Pandas.

## Importación de librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Carga de datos

```python
df = pd.read_csv('../data/raw/Raw Data.csv')
df.head()
```

## 7. Diagnóstico complementario de calidad de datos

```python
# Resumen general por columna
resumen_columnas = pd.DataFrame({
    'tipo_dato': df.dtypes.astype(str),
    'valores_nulos': df.isnull().sum(),
    'porcentaje_nulos': (df.isnull().mean() * 100).round(2),
    'valores_unicos': df.nunique(dropna=True)
})

resumen_columnas
```

```python
# Total de filas duplicadas exactas
duplicados_exactos = df.duplicated().sum()
print("Filas duplicadas exactas:", duplicados_exactos)

# Posibles duplicados lógicos por combinación de nombre, plataforma y año
duplicados_logicos = df.duplicated(subset=['Name', 'Platform', 'Year']).sum()
print("Posibles duplicados por Name + Platform + Year:", duplicados_logicos)
```

```python
# Revisión de posibles inconsistencias entre ventas regionales y ventas globales
suma_regional = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum(axis=1)
diferencia = (suma_regional - df['Global_Sales']).round(2)

print("Máxima diferencia absoluta entre suma regional y Global_Sales:", diferencia.abs().max())
print("Cantidad de filas con diferencia > 0.01:", (diferencia.abs() > 0.01).sum())
print("Cantidad de filas con diferencia > 0.02:", (diferencia.abs() > 0.02).sum())
```

```python
# Frecuencias de algunas variables categóricas
print("Top 10 plataformas:")
display(df['Platform'].value_counts().head(10).to_frame('frecuencia'))

print("\nTop 10 géneros:")
display(df['Genre'].value_counts().head(10).to_frame('frecuencia'))

print("\nTop 10 publishers:")
display(df['Publisher'].value_counts().head(10).to_frame('frecuencia'))
```
