# DataFrames

```python
import pandas as pd
import numpy as np
import sqlite3
```

## Ejemplo básico de creación y manipulación de un DataFrame

```python
# Crear un DataFrame a partir de un diccionario
data = {
    'Nombre': ['Ana', 'Luis', 'Marta'],
    'Edad': [23, 35, 29],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia']
}
df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)
```

```python
# Seleccionar una columna
print(df['Nombre'])

# Filtrar filas donde la edad es mayor de 25
filtered_df = df[df['Edad'] > 25]
print(filtered_df)

# Añadir una nueva columna
df['Salario'] = [50000, 60000, 55000]
print(df)

# Eliminar una columna sin usar inplace=True
df = df.drop(columns=['Ciudad'])
print(df)
```

## Formas de crear DataFrames

```python
# Crear DataFrames a partir de diccionarios

data = {
    'Nombre': ['Ana', 'Luis', 'Marta'],
    'Edad': [23, 35, 29],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia']
}
df = pd.DataFrame(data)
print(df)
```

```python
# Crear DataFrames a partir de listas de listas

data = [
    ['Ana', 23, 'Madrid'],
    ['Luis', 35, 'Barcelona'],
    ['Marta', 29, 'Valencia']
]
df = pd.DataFrame(data, columns=['Nombre', 'Edad', 'Ciudad'])
print(df)
```

```python
# Crear DataFrames a partir de listas de diccionarios

data = [
    {'Nombre': 'Ana', 'Edad': 23, 'Ciudad': 'Madrid'},
    {'Nombre': 'Luis', 'Edad': 35, 'Ciudad': 'Barcelona'},
    {'Nombre': 'Marta', 'Edad': 29, 'Ciudad': 'Valencia'}
]
df = pd.DataFrame(data)
print(df)
```

```python
# Crear DataFrames a partir de arrays de Numpy

data = np.array([
    ['Ana', 23, 'Madrid'],
    ['Luis', 35, 'Barcelona'],
    ['Marta', 29, 'Valencia']
])
df = pd.DataFrame(data, columns=['Nombre', 'Edad', 'Ciudad'])
print(df)
```

```python
# Crear DataFrames a partir de series de Pandas

nombres = pd.Series(['Ana', 'Luis', 'Marta'])
edades = pd.Series([23, 35, 29])
ciudades = pd.Series(['Madrid', 'Barcelona', 'Valencia'])

df = pd.DataFrame({
    'Nombre': nombres,
    'Edad': edades,
    'Ciudad': ciudades
})
print(df)
```

```python
# Crear DataFrames a partir de archivos CSV

df = pd.read_csv('../data/raw/DataFrames.csv')
print(df)
```

```python
# Desde un archivo Excel

df_excel = pd.read_excel('../data/raw/excel.xlsx')
print(df_excel)
```

```python
# Desde un archivo JSON

df_json = pd.read_json('../data/raw/csvjson.json')
print(df_json)
```

```python
# Desde una consulta SQL

conn = sqlite3.connect('../data/raw/database.db')
df_sql = pd.read_sql_query('SELECT * FROM info', conn)
print(df_sql)
```

## Tipos de índices

```python
# Índice numérico secuencial (por defecto)

data = {'Nombre': ['Ana', 'Luis', 'Marta'], 'Edad': [23, 35, 29]}
df = pd.DataFrame(data)
print(df)
```

```python
# Índice basado en una columna existente

df.set_index('Nombre', inplace=True)
print(df)
```

```python
# Índice multi-nivel (MultiIndex)

arrays = [['Ana', 'Ana', 'Luis', 'Luis'], ['Math', 'Science', 'Math', 'Science']]
index = pd.MultiIndex.from_arrays(arrays, names=('Nombre', 'Asignatura'))
data = {'Notas': [90, 85, 88, 92]}
df = pd.DataFrame(data, index=index)
print(df)
```

```python
# Crear un DataFrame de ejemplo con un índice personalizado

data = {
    'Nombre': ['Ana', 'Juan', 'Pedro', 'Laura'],
    'Edad': [23, 25, 30, 28],
    'Ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia']
}
df = pd.DataFrame(data)
df.set_index('Nombre', inplace=True)

print("DataFrame con índice personalizado")
print(df)
```

```python
# Restablecer el índice al valor por defecto

df.reset_index(inplace=True)

print("\nDataFrame después de restablecer el índice:")
print(df)
```

```python
# Renombrar el índice

df.index.name = 'Identificador'
print(df)
```

```python
# Establecer la columna 'Nombre' como índice
df.set_index('Nombre', inplace=True)

# Seleccionar datos mediante el índice
fila_ana = df.loc['Ana']
print(fila_ana)
```

```python
# Crear un índice de fechas

date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
df = pd.DataFrame(date_rng, columns=['Fecha'])
df['Datos'] = range(10)
df.set_index('Fecha', inplace=True)
print(df)
```

```python
# Seleccionar datos de una fecha específica

print(df.loc['2023-01-03'])

# Seleccionar datos de un rango de fechas

print(df.loc['2023-01-03':'2023-01-05'])
```

```python
# Detectar índices duplicados

df = pd.DataFrame({'Nombre': ['Ana', 'Luis', 'Ana'], 'Edad': [23, 35, 29]})
df.set_index('Nombre', inplace=True)
print(df.index.duplicated())

# Eliminar índices duplicados

df = df[~df.index.duplicated(keep='first')]
print(df)
```

```python
# Crear un multiindex

arrays = [['Ana', 'Ana', 'Luis', 'Luis'], ['Math', 'Science', 'Math', 'Science']]
index = pd.MultiIndex.from_arrays(arrays, names=('Nombre', 'Asignatura'))
data = {'Notas': [90, 85, 88, 92]}
df = pd.DataFrame(data, index=index)
print(df)
```

```python
# Seleccionar todas las asignaturas de 'Ana'

print(df.loc['Ana'])

# Seleccionar la nota de 'Luis' en 'Math'

print(df.loc['Luis', 'Math'])
```

## Atributos de DataFrames

```python
# Uso de shape para devolver una tupla que representa las dimensiones del DataFrame, donde el primer elemento es el número de filas y el segundo es el número de columnas.

data = {'Nombre': ['Ana', 'Luis', 'Marta'], 'Edad': [23, 35, 29]}
df = pd.DataFrame(data)
print(df.shape)
```

```python
# Uso de size que devuelve el número total de elementos en el DataFrame (filas multiplicadas por columnas)

print(df.size)
```

```python
# Uso de ndim que devuelve el número de dimensiones del DataFrame, que siempre será 2 para los DataFrames

print(df.ndim)
```

```python
# Uso de columns que devuelve un Index con las etiquetas de las columnas del DataFrame

print(df.columns)
```

```python
# Uso de index que devuelve un Index con las etiquetas de las filas del DataFrame

print(df.index)
```

```python
# Uso de dtypes que devuelve una serie con los tipos de datos de cada columna del DataFrame

print(df.dtypes)
```

```python
# Uso de values que devuelve una representación en forma de array de NumPy de los datos del DataFrame. Aunque el atributo values sigue funcionando, se recomienda utilizar el método to_numpy() por ser más flexible y seguro en ciertas operaciones

print(df.values)
```

```python
# Uso de T que devuelve la transposición del DataFrames, intercambiando filas por columnas

print(df.T)
```

```python
# Uso de head(n) (aunque no es un atributo, es útil mencionarlo) que devuelve las primeras n filas del DataFrame. Por defecto n es 5.

print(df.head(2))
```

```python
# Uso de tail(n) similar a head, este método devuelve las últimas n filas del DataFrame

print(df.tail(2))
```

```python
# Uso de memory_usage(deep=True) nos devuvelve una serie con el uso de memoria de cada columna en bytes. Si deep=True, incluye el uso de memoria de objetos

print(df.memory_usage(deep=True))
```

```python
# Uso de empty que devuelve True si el DataFrame está vacío (No contiene elementos) y False en caso contrario

print(df.empty)
```

## Métodos básicos de DataFrames

```python
# Uso de head(n=5) que devuelve las primeras n filas del DataFrame. Por defecto n es 5

data = {'Nombre': ['Ana', 'Luis', 'Marta', 'Juan', 'Carlos'], 'Edad': [23, 35, 29, 40, 22]}
df = pd.DataFrame(data)
print(df.head(3))
```

```python
# Uso de tail(n=5) que devuelve las últimas n filas del DataFrame. Por defecto, n es 5

print(df.tail(2))
```

```python
# Uso de to_numpy() Este método devuelve los datos del DataFrame como un array de Numpy. Se recomienda utilizar to_numpy() en lugar de values, ya que es más flexible y menos propenso a errores cuando se trabaja con diferentes tipos de datos y estructuras.

print(df.to_numpy())
```

```python
# Uso de info() que proporciona un resumen conciso del DataFrame, incluyendo el índice, el tipo de datos de cada columna y el uso de memoria.

df.info()
```

```python
# Uso de describe() que genera estadísticas descriptivas que resumen la tendencia central, la dispersión y la forma de la distribución de un conjunto de datos, excluyendo valores NaN

print(df.describe())
```

```python
# Uso de mean() que devuelve el promedio de los valores de cada columna numérica

print(df['Edad'].mean())
```

```python
# Uso de sum() que devuelve la suma de los valores de cada columna numérica

print(df['Edad'].sum())
```

```python
# Uso de max() que devuelve el valor máximo de cada columna

print(df['Edad'].max())
```

```python
# Uso de min() que devuelve el valor mínimo de cada columna

print(df['Edad'].min())
```

```python
# Uso de count() que devuelve el número de valores no nulos en cada columna

print(df.count())
```

```python
# Uso de value_counts() que devuelve una serie con la frecuencia de cada valor único en una columna

print(df['Nombre'].value_counts())
```

```python
# Uso de drop(columns) que elimina una o más columnas del DataFrame

df_sin_edad = df.drop(columns=['Edad'])
print(df_sin_edad)
```

```python
# Uso de dropna() que elimina las filas con valores nulos (NaN)

df_con_na = pd.DataFrame({'Nombre': ['Ana', 'Luis', None], 'Edad': [23, None, 29]})
df_sin_na = df_con_na.dropna()
print(df_sin_na)
```

```python
# Uso de fillna(valor) que rellena los valores nulos (NaN) con un valor especificado

df_rellenado = df_con_na.fillna(0)
print(df_rellenado)
```

```python
# Uso de sort_values(by) que ordena el DataFrame por lo valores de una o más columnas

df_ordenado = df.sort_values(by='Edad')
print(df_ordenado)
```

```python
# Uso de groupby(by) que agrupa el DataFrame utilizando una o más columnas y aplica una función agregada a cada grupo

data = {'Nombre': ['Ana', 'Luis', 'Marta', 'Ana', 'Luis'],
       'Edad': [23, 35, 29, 22, 36],
       'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Madrid', 'Barcelona']}
df = pd.DataFrame(data)
df_agrupado = df.groupby('Ciudad').mean(numeric_only=True)
print(df_agrupado)
```

## Acceso a datos del DataFrame con iloc y loc

```python
# Uso de iloc: acceso por posición. Seleccionar una fila específica por su posición

data = {'Nombre': ['Ana', 'Luis', 'Marta'], 'Edad': [23, 35, 29]}
df = pd.DataFrame(data)

# Seleccionar la primera fila
print(df.iloc[0])
```

```python
# Seleccionar un rango de filas

print(df.iloc[0:2])
```

```python
# Seleccionar filas y columnas específicas

print(df.iloc[0, 1])
```

```python
# Seleccionar las dos primeras filas y las dos primeras columnas

print(df.iloc[0:2, 0:2])
```

```python
# Acceso por etiqueta

# Establecer 'Nombre' como índice
df.set_index('Nombre', inplace=True)

# Seleccionar la fila correspondiente a 'Ana'
print(df.loc['Ana'])
```

```python
# Seleccionar las filas correspondientes a 'Ana' y 'Luis'

print(df.loc['Ana':'Luis'])
```

```python
# Seleccionar la fila correspondiente a 'Ana' y la columna 'Edad'

print(df.loc['Ana', 'Edad'])
```

```python
# Seleccionar las filas de 'Ana' a 'Luis' y las columnas 'Edad'

print(df.loc['Ana':'Luis', 'Edad'])
```

```python
# Seleccionar las dos primeras filas y las dos primeras columnas

print(df.iloc[0:2, 0:2])
```

```python
# Restablecer índice para usar etiquetas

df.reset_index(inplace=True)

# Seleccionar filas por etiquetas y columnas por nombres

print(df.loc[0:1, ['Nombre', 'Edad']])
```

```python
# Filtrar filas donde la edad es mayor de 

print(df.loc[df['Edad'] > 25])
```

```python
# Seleccionar filas donde la edad es mayor de 25 y el nombre es 'Luis'

print(df.loc[(df['Edad'] > 25) & (df['Nombre'] == 'Luis')])
```

## ¿Cómo contribuye el uso de DataFrame en Pandas a la eficiencia y comprensión de los datos en un proyecto de análisis?

El uso de DataFrames en Pandas contribuye significativamente a la eficiencia y comprensión de los datos en un proyecto de análisis, ya que permite organizar la información en una estructura tabular clara y fácil de interpretar. Gracias a sus funciones, es posible realizar procesos como limpieza, transformación y análisis de datos de manera rápida y optimizada, evitando el uso de métodos más complejos o lentos. Además, facilita la exploración de los datos mediante estadísticas y filtros que ayudan a identificar patrones, errores o tendencias. En conjunto, el DataFrame se convierte en una herramienta fundamental que no solo agiliza el trabajo del analista, sino que también mejora la claridad de los resultados obtenidos.
